# Dropping duplicates + empty cells

```python
import numpy as np
import pandas as pd

df = pd.read_csv("may14generated.txt", header=None)
df
```

```python
df[0][17680]
```

```python
df = df.drop_duplicates()
df
```

```python
df.info()
```

```python
df = df.reset_index(drop=True)

df
```

```python
df = df.dropna()
df
```

```python
import numpy as np

df1 = df.replace(r'^\s*$', np.nan, regex=True)
df1 = df.replace("NaN", np.nan)
df1 = df.dropna()
df1 = df1.reset_index(drop=True)

df1
```

# Deleting non-recognized-as-English entries

```python
from langdetect import detect

lang = detect(df[0].iloc[2])

print(lang)
```

```python
df1["language"] = ""
lol = [detect(elem) for elem in df1[0]]
```

```python
df1["language"] = lol
df1
```

```python
df1 = df1[(df1['language']=='en')]

df1
```

```python
df1 = df1.reset_index(drop=True)

df1
```

# Spell checking the entries
