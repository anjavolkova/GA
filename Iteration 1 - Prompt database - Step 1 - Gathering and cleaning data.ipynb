# Removing popular words and symbols
Some of the common words associated with music (Acoustic, feat., Remix...) and symbols negatively affected the quality of the generated prompts
- deleing na
- deleting duplicates

```python
df = df["song_name"].dropna()
df = df.drop_duplicates()
df = df.reset_index(drop=True)
df
```

```python
from collections import Counter

string_titles=df.to_string()
```

```python
most_occur = Counter(string_titles.split()).most_common(100)
most_occur
```

```python
SYMBOLS = ["   !",'(with','(feat.','Remix','Version',"'","{","}","(",")","[","]",".",",",':',";",'+','-','*','/','&','|','<','>','=','~','$','1',"2",'3','4','5','6','7','8','9','0','#',".", 'Remastered','Radio Edit','Acoustic']

df = df.astype(str)

for i in range(len(df)):
    for symbol in SYMBOLS:
        df.iloc[i] = df.iloc[i].replace(symbol, '')
    
df.head()
```

```python
df = df.dropna()
df = df.drop_duplicates()

df
```

# Detecting language
Non-english entries in the database negatively affected the quality of the generated text as well

```python
import numpy as np

df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.replace("NaN", np.nan)
df = df.dropna()
df = df.reset_index(drop=True)

df
```

```python
from langdetect import detect

lang = detect(df.iloc[12000])

print(lang)
```

```python
for i in range(len(df)):
    try:
        language = detect(df.iloc[i])
    except:
        language = "error"
        print("This row throws an error:", df.iloc[i])
```

```python
df = df[df != "  !"]
df = df[df != '""']
df
```

```python
langlist = []
langlist = [detect(elem) for elem in df]
langlist
```

```python
df = df.tolist()
```

```python
df = pd.DataFrame(list(zip(df, langlist)),
               columns =['title', 'lang'])
df
```

```python
df = df[(df['lang']=='en')]

df
```
