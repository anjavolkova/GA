## Running the API:
Because of rate limits, code was run multiple times on smaller sections, and the sections were added to the database.

In this case, the code was run on 815 lines before hitting the rate limit, so the first 815 titles + the colors were added to a database

```python
import time
colors1 = []


#https://medium.com/nerd-for-tech/create-ai-application-in-minutes-with-openai-api-5e84bd3ec5d0
#https://www.seotraininglondon.org/free-gpt3-colour-palette-generator-v1/

for i in range(len(titles)):
    time.sleep(10)
    description = titles[i]
    response = openai.Completion.create(
        engine="davinci",
        prompt= data + description + "\nHex Values:",
        temperature=0.3,
        max_tokens=64,
        top_p=1,
        frequency_penalty=0.5,
        presence_penalty=0,
        stop=["###\n"])
    colors1.append(response["choices"][0]["text"])
        
colors1
```

## Creating the matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df1['soup'])
```

```python
count_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
indices = pd.Series(df1.index, index=df1['Title'])
```

## Creating thumbs up and down functions

```python
def thumbs_up(title, cosine_sim2=cosine_sim2):
    
    index = indices[title]

    sim_scores = list(enumerate(cosine_sim2[index]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[0:15] 

    prompts = [i[0] for i in sim_scores]

    return df1.iloc[prompts]
```

```python
def thumbs_down(title, cosine_sim2=cosine_sim2):

    index = indices[title]

    sim_scores = list(enumerate(cosine_sim2[index]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[200:210]

    prompts = [i[0] for i in sim_scores]

    return df1.iloc[prompts]
```
